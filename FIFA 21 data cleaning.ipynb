```python
def check_contract(val):
    if 'On Loan' in val:
        return 'On Loan'
    elif 'Free' in val:
        return 'Free Agent'
    elif '~' in val:
        return 'Under Contract'
    else:
        return 'Unknown'
```

```python
fifa1d['Contract Term'] = fifa1d['Contract Term'].apply(check_contract)
```

### Convert the values of HEIGHT column to (cm)

```python
inches = fifa1d[fifa1d.Height.str.contains("'")]
inches_height = inches['Height']
inches_height.sample(5)
```

```python
def inches_to_cm(val):
    if "'" in val:
        val = val.strip('"') #removes (") from the end of all inches value
        feet,inches = val.split("'") # splits each value into feet and inches 
        inches_val = round((float(feet)*30.48 + float(inches)*2.54),2) # Calculation to convert 'ft/in' to 'cm'
        return inches_val
    else:
        val = val.strip('cm') # removes 'cm' label from cm values
        return val
    
fifa1d['Height_cm'] = fifa1d['Height'].apply(lambda x: inches_to_cm(x))
```

```python
#drop previous height column

fifa1d.drop('Height', axis=1, inplace=True)
fifa1d[['Height_cm']].head()
```

### Change values of WEIGHT column to (kg)

```python
def lbs_to_kg (val):
    if 'lbs' in val:
        val = val.strip('lbs')
        kg_val = round((float(val)*0.453592),2)
        return kg_val
    else:
        val = val.strip('kg')
        return val
    
fifa1d['weight_kg'] = fifa1d['Weight'].apply(lambda x : lbs_to_kg(x))
fifa1d['weight_kg'].sample(10)

#drop previous weight column
fifa1d.drop('Weight', axis=1, inplace=True)
```

### Remove currency symbol from VALUE and RELEASE CLAUSE

## Calculate percentage value for summed-up columns 


Calculate the percentage value for each row in the summed columns.

To determine the percentage value for each row of the summed columns (Attacking, Skills, Movement, Power, Mentality, Defending, and Goalkeeping), follow these steps:

1. Identify the columns in question, for example, Attacking, which comprises Crossing, Finishing, Heading Accuracy, Short Passing, and Volleys.

2. Find the highest possible value for these columns. In this example, the highest value for each attribute (Crossing, Finishing, Heading Accuracy, Short Passing, Volleys) is set to 100. Therefore, the collective highest possible value for Attacking is 500 (100 * 5).

3. For each row in the Attacking column, divide the value by the collective highest possible value (500) and then multiply by 100. This calculation yields the percentage value for that particular row, indicating the contribution of Attacking attributes to the overall performance.

The same process can be applied to calculate the percentage values for other summed columns like Skills, Movement, Power, Mentality, Defending, and Goalkeeping.



### Attacking 

The sum of the following columns; Crossing, Finishing, Heading accuracy, Short Passing, Volleys

```python
def attacking(val):
    val = (val/500) * 100
    return val

fifa1d['Attacking_percentage'] = ['Attacking'].apply(lambda x: attacking(x))
```

## Skills

```python
def skill(val):
    val = (val/500) * 100
    return val

fifa1d['Skill_percentage'] = fifa1d['Skill'].apply(lambda x: skill(x))
```
